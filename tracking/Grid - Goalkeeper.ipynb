```python
# Version 1 : Grid accross the entire image

import pylab as plt

# Load the image
img = plt.imread("img/goalkeeper.png")

# Grid lines at these intervals (in pixels)
# dx and dy can be different
dx, dy = 100,100

# Custom (rgb) grid color
grid_color = [0,0,0]

# Modify the image to include the grid
img[:,::dy,:] = grid_color
img[::dx,:,:] = grid_color

# Show the result
plt.imshow(img)
plt.show()
```

```python
# Version 2 : Grid accross the goal only using coordinate points

import pylab as plt

# Load the image
img = plt.imread("img/goalkeeper.png")

# Designate points
top_left = (100,100)
bot_right = (500,850)

# Grid lines at these intervals (in pixels)
# dx and dy can be different
# dx, dy = 100,100
dx = int((bot_right[0] - top_left[0])/5)
dy = int((bot_right[1] - top_left[1])/5)

print(dx,dy)
# Custom (rgb) grid color
grid_color = [0,0,0]

# Modify the image to include the grid
img[top_left[0]:bot_right[0] ,top_left[1]::dy,:] = grid_color
img[top_left[0]::dx,top_left[1]:bot_right[1],:] = grid_color

# Show the result
plt.imshow(img)
plt.show()
```

```python
# Version 3 : Grid using user input to place points
from tkinter import *
# pip install pillow
import pylab as plt

event2canvas = lambda e, c: (c.canvasx(e.x), c.canvasy(e.y))

root = Tk()

#setting up a tkinter canvas with scrollbars
frame = Frame(root, bd=2, relief=SUNKEN)
frame.grid_rowconfigure(0, weight=1)
frame.grid_columnconfigure(0, weight=1)
xscroll = Scrollbar(frame, orient=HORIZONTAL)
xscroll.grid(row=1, column=0, sticky=E+W)
yscroll = Scrollbar(frame)
yscroll.grid(row=0, column=1, sticky=N+S)
canvas = Canvas(frame, bd=0, xscrollcommand=xscroll.set, yscrollcommand=yscroll.set)
canvas.grid(row=0, column=0, sticky=N+S+E+W)
xscroll.config(command=canvas.xview)
yscroll.config(command=canvas.yview)
frame.pack(fill=BOTH,expand=1)

#adding the image
img_tk = PhotoImage(file="img/goalkeeper.png")
canvas.create_image(0,0,image=img_tk,anchor="nw")
canvas.config(scrollregion=canvas.bbox(ALL))

pos_points = []

#function to be called when mouse is clicked
def printcoords(event):
    global pos_points
    #outputting x and y coords to console
    cx, cy = event2canvas(event, canvas)
    pos_points.append((int(cy),int(cx)))
    print ("(%d, %d) / (%d, %d)" % (event.y,event.x,cy,cx))
    if (len(pos_points) == 2):
        root.destroy()
#mouseclick event
canvas.bind("<ButtonPress-1>",printcoords)

root.mainloop()
```
